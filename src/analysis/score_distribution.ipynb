```python
% matplotlib inline

import pandas as pd
from definitions import *
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.python.keras.preprocessing.text import Tokenizer

sentences_file = 'AmazonReviews.csv'
BASE_DIR = os.path.join(ROOT_DIR, 'data')
sentences_path = os.path.join(BASE_DIR, sentences_file)
```

```python
scores = pd.read_csv(sentences_path, encoding='utf8', quoting=csv.QUOTE_NONE, usecols=['Score'])['Score']

```

```python
counts = scores.value_counts(sort=False)
```

```python
# counts[1:6].plot(kind='bar' )
plt.bar([1,2,3,4,5],counts[1:6])
plt.savefig("original_scores_distribution.pdf")
```
